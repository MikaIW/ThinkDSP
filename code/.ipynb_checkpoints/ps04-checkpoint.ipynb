### Phase Offsets

1) Please finish this exercise that we started in class: One of the examples in chap07.ipynb compares two mixtures of complex sinusoids, one with offset=0 and one with offset=1.  The result is a different looking waveform, because "rotating" each component by the same phase offset does not shift each component by the same amount of time.

```python
amps = numpy.array([0.6, 0.25, 0.1, 0.05])
freqs = [100, 200, 300, 400]
framerate = 11025
ts = numpy.linspace(0, 1, framerate)
```

```python
def synthesize2(amps, freqs, ts):
    args = numpy.outer(ts, freqs)
    M = numpy.exp(i * PI2 * args)
    ys = numpy.dot(M, amps)
    return ys

ys = synthesize2(amps, freqs, ts)
wave = thinkdsp.Wave(ys.real, framerate)

amps2 = amps * numpy.exp(i)
ys2 = synthesize2(amps2, freqs, ts)

n=400
thinkplot.plot(ts[:n], ys.real[:n], label=r'$\phi_0 = 0$')
thinkplot.plot(ts[:n], ys2.real[:n], label=r'$\phi_0 = 1$')
thinkplot.config(ylim=[-1.15, 1.05])
```

The differences in phase offset above produce waves of different shapes, rather than having identical waves shifted in the time domain. This is because amps2 is only rotating the wave at one frequency per period, which ultimately distorts the wave.

So to shift the entire wave in the time domain while retaining the same shape, amps2 needs to take into account frequency in its exponent. This applies the phase shift to every frequency of the wave.

```python
def synthesize2(amps, freqs, ts):
    args = numpy.outer(ts, freqs)
    M = numpy.exp(i * PI2 * args)
    ys = numpy.dot(M, amps)
    return ys

ys = synthesize2(amps, freqs, ts)
wave = thinkdsp.Wave(ys.real, framerate)

amps2 = amps * numpy.exp(i*numpy.array(freqs)) # new amps2
ys2 = synthesize2(amps2, freqs, ts)

n=400
thinkplot.plot(ts[:n], ys.real[:n], label=r'$\phi_0 = 0$')
thinkplot.plot(ts[:n], ys2.real[:n], label=r'$\phi_0 = 1$')
thinkplot.config(ylim=[-1.15, 1.05])
```

### Discrete Fourier Transform

#### Allen's DFT:

```python
def dft(ys):
    N = len(ys)
    ts = numpy.arange(N) / N
    freqs = numpy.arange(N)
    args = numpy.outer(ts, freqs)
    M = numpy.exp(i * PI2 * args)
    amps = M.conj().transpose().dot(ys)
    return amps
```

#### My Wikipedia-Based DFT:

My new DFT function computes DFT using two for loops. The inner for loop loops through n values of the input array and computes the inside of the summation, $x_n e^{\frac{-2\pi ink}{N}}$. The outside loop places each one of k values of X into the output array, $X_k$.

```python
def mynewdft(x):
    """Wikipedia discrete Fourier Transform"""
    N = len(x)
    X = numpy.zeros(N,dtype=complex)
    for k in range(N):
        for n in range(N):
            M = numpy.exp(-i * PI2 * k * n / N)
            X[k] += x[n]*M
    return X
```

#### Test the Results:

```python
test = numpy.array([0.6, 0.25, 0.1, 0.05*i])
```

```python
print(mynewdft(test))
```

```python
print(dft(test))
```

The results of my summation-based DFT matched the results of Allen's matrix multiplication-based DFT.

### Cooley-Tukey Fast Fourier Transpose

```python
def is_power2(n):
    return n != 0 and ((n & (n-1)) == 0)

def FFT(x):
    """Cooley-Tukey FFT algorithm, with help from Kyle Flores!"""
    N = len(x)
    
    if is_power2(N) == 0:
        raise Exception("length of x must be a power of 2")
    if N <= 1:
        return x # base case
    else:
        X_even = FFT(x[::2]) # list of every two elements (all the evens) of the fft of x
        X_odd = FFT(x[1::2]) # list of every two elements (all the odds) of the fft of x
        
        X = numpy.zeros(N,dtype=complex)
        twiddle = numpy.zeros(N,dtype=complex)
        
        for k in range(N):
            twiddle[k] = numpy.exp(-i*PI2*k/N)
            m = k%(N/2) # index the half-size arrays
            X[k] = X_even[m] + twiddle[k]*X_odd[m]
        
    return X
```

```python
print(FFT(test))
```

The Fast Fourier Transform computes the DFT in a more efficient way by first splitting up the DFT values into its even- and odd-numbered values, then using the symmetries of the even and odd terms to further split up the DFT computation and reduce computation time. The FFT results in a computation time of scale O[NlogN], rather than DFT's O[$N^2$].

### Correlation Timing

#### autocorr:

```python
def make_wave(offset):
    signal = thinkdsp.SinSignal(freq=440, offset=offset)
    wave = signal.make_wave(duration=0.5, framerate=10000)
    return wave

def compute_corr(offset):
    wave1 = make_wave(0)
    wave2 = make_wave(offset=-offset)
    
    thinkplot.preplot(2)
    wave1.segment(duration=0.01).plot()
    wave2.segment(duration=0.01).plot()
    
    corr = wave1.corr(wave2)
    print('corr =', corr)
    
    thinkplot.config(xlabel='time (s)', legend=False, ylim=[-1.05, 1.05])
    
def corrcoef(xs, ys):
    return numpy.corrcoef(xs, ys, ddof=0)[0, 1]

def serial_corr(wave, lag=1):
    N = len(wave)
    y1 = wave.ys[lag:]
    y2 = wave.ys[:N-lag]
    corr = corrcoef(y1, y2)
    return corr

def autocorr(wave):
    """Computes and plots the autocorrelation function.

    wave: Wave
    """
    lags = range(len(wave.ys)//2)
    corrs = [serial_corr(wave, lag) for lag in lags]
    return lags, corrs
```

```python
signal = thinkdsp.UncorrelatedGaussianNoise()
wave = signal.make_wave(duration=100)

ns = 2 ** numpy.arange(0, 14)

results = []
for N in ns:
    ys = thinkdsp.Wave(wave.ys[:N],framerate)
    result = %timeit -o autocorr(ys)
    results.append(result)
```

```python
bests = [result.best for result in results]
bests
```

```python
thinkplot.plot(ns, bests)
thinkplot.config(xscale='log', yscale='log')
```

We can see from the +1 slope of the computation time above that autocorr has a computation time with the scale O[N].

#### fft_autocorr:

```python
def zero_pad(array, n):
    res = numpy.zeros(n)
    res[:len(array)] = array
    return res

def fft_convolve(signal, window):
    fft_signal = numpy.fft.fft(signal)
    fft_window = numpy.fft.fft(window)
    return numpy.fft.ifft(fft_signal * fft_window)

def fft_autocorr(signal):
    N = len(signal)
    window = signal[::-1]
    signal = zero_pad(signal, 2*N)
    window = zero_pad(window, 2*N)

    corrs = fft_convolve(signal, window)
    corrs = corrs[N//2: 3*N//2]
    return corrs
```

```python
results2 = []

for N in ns:
    ys = wave.ys[:N]
    result2 = %timeit -o fft_autocorr(ys)
    results2.append(result2)
```

```python
bests2 = [result2.best for result2 in results2]
bests2
```

```python
thinkplot.plot(ns, bests2)
thinkplot.config(xscale='log', yscale='log')
```

The computation time of fft_autocorr creates a plot with an exponential slope. This makes sense because the computation time of fft_autocorr has an O[NlogN] scale.
