```python
from __future__ import print_function, division

import thinkdsp
import thinkplot
import numpy
%matplotlib inline
```

```python
filename = '193170__netaj__thunder.wav' #thunder .wav file
bells = thinkdsp.read_wave(filename)
bells.make_audio()
```

### Wave Segment

```python
start = 0
duration = 0.8
segment = bells.segment(start, duration)
segment.plot()
```

Because this is a segment of a thunderstorm sound, I don't see any repeated sinusoid periods. Instead, I have plotted noise.

### Spectrum Plot

```python
spectrum = segment.make_spectrum()
spectrum.plot()
```

There doesn't seem to be any recognizable harmonic structure, again due to the fact that this is a thunder noise, and particular notes aren't being played.

```python
spectrum.make_wave().make_audio()
```

#### Low Pass Filter

```python
spectrum.low_pass(3000)
spectrum.plot()
```

```python
spectrum.make_wave().make_audio()
```

#### High Pass Filter

```python
segment = bells.segment(start, duration)
spectrum = segment.make_spectrum()
spectrum.high_pass(10000)
spectrum.plot()
```

```python
spectrum.make_wave().make_audio()
```

#### Band Pass Filter

```python
segment = bells.segment(start, duration)
spectrum = segment.make_spectrum()
spectrum.band_stop(4000,16000)
spectrum.plot()
```

```python
spectrum.make_wave().make_audio()
```

### Synthesizing a Wave

#### Signals that are Multiples of the Fundamental:

```python
cos_sig = thinkdsp.CosSignal(freq=440, amp=1.0, offset=0)
sin_sig = thinkdsp.SinSignal(freq=880, amp=0.5, offset=0)
cos_sig2 = thinkdsp.CosSignal(freq=1320, amp=0.75, offset=0)
```

#### Not Multiples of the Fundamental:

```python
sin_sig2 = thinkdsp.SinSignal(freq=1000, amp=1.0, offset=0)
cos_sig3 = thinkdsp.CosSignal(freq=1200, amp=0.5, offset=0)
```

#### Summing the Signals with Multiples

```python
mix = sin_sig + cos_sig + cos_sig2
mywave = mix.make_wave(duration=1.5, start=0, framerate=11025)
mywave
```

```python
period = mix.period
segment2 = mywave.segment(start=0, duration=period*4)
segment2.plot()
```

#### Summing all Signals:

```python
mix2 = sin_sig + cos_sig + cos_sig2 + sin_sig2 + cos_sig3
mywave2 = mix2.make_wave(duration=1.5, start=0, framerate=11025)
period = mix2.period
segment3 = mywave2.segment(start=0, duration=period*4)
segment3.plot()
```

When I added frequencies that weren't multiples of the fundamental, the wave looked much less periodic and the pattern was lost.

### Sawtooth Signal

```python
class MySawtoothSignal(thinkdsp.Sinusoid):
    
    def evaluate(self,ts):
        cycles = self.freq * ts + self.offset / thinkdsp.PI2
        frac, _ = numpy.modf(cycles)
        ys = thinkdsp.normalize(thinkdsp.unbias(frac), self.amp) #scales it to the amplitude
        return ys
```

```python
MySawtoothSignal(freq=440, amp=1.0, offset=0).plot()
```

```python
sawwave = MySawtoothSignal(freq=440, amp=1.0, offset=0).make_wave(duration = 5.0, framerate = 50000)
sawspectrum = sawwave.make_spectrum()
sawspectrum.plot()
```

#### Triangle Wave

```python
triwave = thinkdsp.TriangleSignal(freq=440, amp=1.0, offset=0).make_wave(duration = 5.0 ,framerate = 50000)
trispectrum = triwave.make_spectrum()
trispectrum.plot()
```

The triangle wave drops off much quicker than the sawtooth wave (it behaves like 1/f^2) and has half as many harmonics, because it only includes odd harmonics, while the sawtooth wave includes both odd and even harmonics.

#### Square Wave

```python
squarewave = thinkdsp.SquareSignal(freq=440, amp=1.0, offset=0).make_wave(duration = 5.0 ,framerate = 50000)
squarespectrum = squarewave.make_spectrum()
squarespectrum.plot()
```

The square wave drops off like the sawtooth wave (it behaves like 1/f) and also has half as many harmonics, because it only includes odd harmonics.

### 1100 Hz Frequency at 10,000 frames/s

#### Triangle

```python
triwave = thinkdsp.TriangleSignal(freq=1100, amp=1.0, offset=0).make_wave(duration = 5.0 ,framerate = 10000)
triwave.make_audio()
```

The aliasing of the triangle signal can be heard as slightly unclear and more fuzzy than a signal with a the full framerate.

#### Square

```python
squarewave = thinkdsp.SquareSignal(freq=1100, amp=1.0, offset=0).make_wave(duration = 5.0 ,framerate = 10000)
squarewave.make_audio()
```

I could hear much more aliasing in the square wave than in the triangle wave. I think this is because the triangle wave cuts off more quickly, so you mostly hear the fundamental frequency. But in the square wave, since it doesn't cut off as quickly you can hear a lot more of the noise in between and it becomes more discordant.

### Glottal Signal

```python
glottalwave = thinkdsp.GlottalSignal(freq=440, amp=1.0, offset=0).make_wave(duration = 5.0, framerate = 50000)
glottalspectrum = glottalwave.make_spectrum()
glottalspectrum.plot()
```

The glottal signal drops off like the triangle signal does, but also includes the even harmonics.
