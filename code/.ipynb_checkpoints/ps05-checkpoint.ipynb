### Synthesis Matrix

1) Following the example in chap07.ipynb, make a synthesis matrix (the IDFT matrix) using a different range of frequencies, and see what difference it makes.

The usual version of IDFT uses frequencies 0 through N-1.  What happens if you go from N to 2N-1?

```python
def dft(ys):
    N = len(ys)
    ts = numpy.arange(N) / N
    freqs = numpy.arange(N)
    args = numpy.outer(ts, freqs)
    M = numpy.exp(i * PI2 * args)
    amps = M.conj().transpose().dot(ys)
    return amps
```

```python
framerate = 10000
signal = thinkdsp.SawtoothSignal(freq=500)
wave = signal.make_wave(duration=0.1, framerate=framerate)
hs = dft(wave.ys)
amps = numpy.absolute(hs)
N = len(wave.ys)
fs = numpy.arange(N)*framerate/N

thinkplot.plot(fs, amps)
thinkplot.config(xlabel='frequency (Hz)', ylabel='amplitude')
```

### Windows Corresponding to Filters

2) In thinkdsp.py, the Spectrum methods low_pass and high_pass are ideal filters, in some sense: in the passband they are perfectly flat, and in the stopband they attenuate to 0 (or whatever value you want).

Invoking the principle that for every filter there is a corresponding window, see if you can figure out what window corresponds to these ideal filters.

#### Square Signal:

I will apply the low-pass and high-pass behaviors to a 440 Hz square wave spectrum.

```python
squarewave = thinkdsp.SquareSignal(freq=440, amp=1.0, offset=0).make_wave(duration = 5.0 ,framerate = 50000)
squarespectrum = squarewave.make_spectrum()
squarespectrum.plot()
```

#### Low-Pass Behavior:

##### Low-Pass Filter:

Below, I have applied a low pass filter to the square signal for comparison:

```python
squarespectrum.low_pass(5000)
squarespectrum.plot()
```

##### Smoothing Window:

Recreating that square wave:

```python
squarewave = thinkdsp.SquareSignal(freq=440, amp=1.0, offset=0).make_wave(duration = 5.0 ,framerate = 50000)
squarespectrum = squarewave.make_spectrum()
```

A smoothing window in the time domain corresponds to a low-pass filter in the frequency domain.

```python
smoothingwindow = numpy.ones(11)
smoothingwindow /= sum(smoothingwindow)
smoothys = numpy.convolve(squarewave.ys, smoothingwindow, mode='same')
smooth = thinkdsp.Wave(smoothys, framerate=squarewave.framerate)

squarespectrum2 = smooth.make_spectrum()
```

```python
squarespectrum2.plot()
squarespectrum.plot(color='0.7')
```

Applying a smoothing window to the square wave then computing its spectrum has a similar effect to the spectrum as applying a low-pass filter to the square wave spectrum. The smoothing window attenuates more seamlessly, while the low-pass filter had a more abrupt cut-off.

#### High Pass Window:

I have now applied a high pass filter to the square signal for comparison. There is some aliasing present in the spectrum plot below, but I'm going to ignore that.

```python
squarespectrum.high_pass(15000)
squarespectrum.plot()
```

Recreating the square wave again:

```python
squarewave = thinkdsp.SquareSignal(freq=440, amp=1.0, offset=0).make_wave(duration = 5.0 ,framerate = 50000)
squarespectrum = squarewave.make_spectrum()
squarespectrum.plot()
```

```python
diff = numpy.diff(squarewave.ys)
diffwave = thinkdsp.Wave(diff,framerate = 50000)
diffspec = diffwave.make_spectrum()
```

```python
diffspec.plot()
```

Computing the difference between successive elements (differentiation), in the time domain, corresponds to a high-pass filter.

### Second Order Differentiation

4) Taking the difference of successive elements in a wave array has the same effect as convolution with the window [1, -1].  The filter for this
window is an increasing function of f, so you can think of diff as a high-pass filter.

One way to explain this behavior is that diff is a rough approximation of a derivative, and differentiation corresponds to a filter where amplitude is proportional to f.

If we use a better approximation of a derivative, we might expect the corresponding filter to be a better approximation of a line.  The window [-1/2, 2, -3/2] is a second order forward difference approximation of the first derivative.  Compute its DFT and see what the filter looks like.

Also try [1, -2, 1], which is approximates the second derivative.  What does the filter for the second derivative look like?  And how does that compare to the filter for this window?

## Derivative of a Parabolic Signal

6) Use spectrum.differentiate() to compute the derivative of a ParabolicSignal.  Hint: this works best if the duration of the wave is an integer number of periods.

```python
parsig = thinkdsp.ParabolicSignal()
parwave = parsig.make_wave()
parseg = parwave.segment(duration = parsig.period * 10)
parseg.plot()
```

```python
parspectrum = parseg.make_spectrum()
parspectrum.plot()
```

```python
parspectrum.differentiate()
newparwave = parspectrum.make_wave()
newparwave.plot()
thinkplot.config(xlim =[0,0.01])
```

```python
newparspectrum = newparwave.make_spectrum()
newparspectrum.plot()
```
