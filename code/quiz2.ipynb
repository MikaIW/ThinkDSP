```python
from __future__ import print_function, division

import thinkdsp
import thinkplot
import numpy

%precision 3
%matplotlib inline
```

```python
class TriangleChirp(thinkdsp.Chirp):

    def _evaluate(self,ts,freqs):
        dts = numpy.diff(ts)
        dps = thinkdsp.PI2 * freqs * dts
        phases = numpy.cumsum(dps)
        phases = numpy.insert(phases, 0, 0)
        
        cycles = phases / thinkdsp.PI2
        frac, _ = numpy.modf(cycles)
        ys = numpy.abs(frac - 0.5)
        ys = thinkdsp.normalize(thinkdsp.unbias(ys), self.amp)
        
        return ys
```

```python
signal = TriangleChirp()
wave = signal.make_wave()
wave.make_audio()
```

```python
TriangleChirp(start=1000,end=1500).make_wave(framerate = 10000).make_spectrum().plot()
```
