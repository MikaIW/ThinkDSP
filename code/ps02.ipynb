```python
from __future__ import print_function, division

import thinkdsp
import thinkplot
import numpy

%precision 3
%matplotlib inline
```

```python
signal = thinkdsp.SinSignal(freq=440)
duration = signal.period * 30.5
wave = signal.make_wave(duration)
wave.plot()
thinkplot.config(xlabel='time (s)', ylabel='amplitude', legend=False)
```

#### Spectrum of a Non-Periodic Segment:

```python
spectrum = wave.make_spectrum()
spectrum.plot(high=50)
thinkplot.config(xlabel='frequency (Hz)', ylabel='amplitude', legend=False)
```

### Spectrum of a Windowed Segment:

#### Bartlett

```python
signal = thinkdsp.SinSignal(freq=440)
wave1 = signal.make_wave(duration)
wave1.window(numpy.bartlett(len(wave1.ys))) #bartlett will restructure the wave's ys
bartlettspectrum = wave1.make_spectrum()
bartlettspectrum.plot()
```

#### Blackman

```python
signal = thinkdsp.SinSignal(freq=440)
wave2 = signal.make_wave(duration)
wave2.window(numpy.blackman(len(wave2.ys)))
blackspectrum = wave2.make_spectrum()
blackspectrum.plot()
```

#### Hanning

```python
signal = thinkdsp.SinSignal(freq=440)
wave3 = signal.make_wave(duration)
wave3.window(numpy.hanning(len(wave3.ys)))
hanningspectrum = wave3.make_spectrum()
hanningspectrum.plot()
```

#### Kaiser

```python
signal = thinkdsp.SinSignal(freq=440)
wave4 = signal.make_wave(duration)
wave4.window(numpy.kaiser(len(wave4.ys),6)) #beta of 6 generates a shape similar to hanning
kaiserspectrum = wave4.make_spectrum()
kaiserspectrum.plot()
```

### Sawtooth Chirp

```python
class SawtoothChirp(thinkdsp.Chirp):

    def _evaluate(self,ts,freqs):
        dts = numpy.diff(ts)
        dps = thinkdsp.PI2 * freqs * dts
        phases = numpy.cumsum(dps)
        phases = numpy.insert(phases, 0, 0)
        
        #replace the ys in Chirp with Sawtooth Signal:
        cycles = phases / thinkdsp.PI2
        frac, _ = numpy.modf(cycles)
        ys = thinkdsp.normalize(thinkdsp.unbias(frac), self.amp)
        
        return ys
```

```python
signal = SawtoothChirp()
wave = signal.make_wave()
wave.make_audio()
```

```python
wave.make_spectrogram(seg_length = 512).plot(high=500)
```

### Vowel Sounds

Make or find a recording of a series of vowel sounds and look at the spectrogram. Can you identify different vowels?

```python
filename1 = 'vowels.wav'
vowelswave = thinkdsp.read_wave(filename1)
vowelswave.make_audio()
```

```python
thinkdsp.read_wave(filename1).make_spectrogram(seg_length = 6).plot()
thinkplot.config(xlabel='time (s)',
                 ylabel='frequency (Hz)')
```

The vowels look distinctly different. 'A' and 'O' are denser in the plot, especially in the lower frequencies, which shows that those vowels hit these low frequencies more frequently than 'E', 'I' or 'U' do.

### A Soft Murmur - Quartz Crystal Singing Bowl

#### Singing Bowl

```python
singingwave = thinkdsp.read_wave('singingbowl.wav')
singingsegment = singingwave.segment(start=44, duration=8)
singingsegment.make_audio()
```

```python
singingspectrum = singingsegment.make_spectrum()
#intsingingspectrum = singingspectrum.make_integrated_spectrum()
singingspectrum.plot_power()
thinkplot.config(xlabel='frequency (Hz)',
                 ylabel='power density',
                 xscale = 'log',
                 yscale = 'log')
```

The power spectrum for the singing bowl sound best resembles pink noise, with a slope of approximately -1, as the relationship between frequency and power for pink noise is P = K/f.

### Bartlett's Method

```python
singingwave = thinkdsp.read_wave('singingbowl.wav')

# 5 Non-Overlapping Data Segments:

ssegment1 = singingwave.segment(start=45, duration=5)
ssegment2 = singingwave.segment(start=50, duration=5)
ssegment3 = singingwave.segment(start=55, duration=5)
ssegment4 = singingwave.segment(start=60, duration=5)
ssegment5 = singingwave.segment(start=65, duration=5)

# Squared Segments:

ssegment1squared = ssegment1*ssegment1
ssegment2squared = ssegment2*ssegment2
ssegment3squared = ssegment3*ssegment3
ssegment4squared = ssegment4*ssegment4
ssegment5squared = ssegment5*ssegment5

# Spectrums:

sspec1 = ssegment1squared.make_spectrum()
sspec2 = ssegment2squared.make_spectrum()
sspec3 = ssegment3squared.make_spectrum()
sspec4 = ssegment4squared.make_spectrum()
sspec5 = ssegment5squared.make_spectrum()

choir = sspec1 + sspec2 + sspec3 + sspec4 + sspec5
choir.hs = choir.hs/5

choir.plot_power()
thinkplot.config(xlabel='frequency (Hz)',
                 ylabel='power density',
                 xscale = 'log',
                 yscale = 'log')
```

The Bartlett Method achieved its goal of decreasing the variance of the periodograms. This plot's power density range is from approximately 10^8 to 10^14, which is comparitively low next to the original plot's range of 10^2 to 10^18.
